```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
from numpy import pi
from skimage import io
from numpy import r_
from google.colab.patches import cv2_imshow
from scipy.fftpack import dct, idct
import cv2 as cv
import numpy as image_3d
import matplotlib.pylab as plt
import matplotlib.pylab as pylab
```

```python
def dct2(a):
    return dct(dct(a.T, norm='ortho').T, norm='ortho')

def idct2(a):
    return idct(idct(a.T, norm='ortho').T, norm='ortho')
```

```python
url = "https://photo-works.net/images/tutorial/how-ot-make-picture-black-and-white-after.jpg"
image = io.imread(url)
#show the image
cv2_imshow(image)
#show the matrix
print (image)
```

```python
imsize = image.shape
dct = np.zeros(imsize)
#applay DCT on 8X8 
for i in r_[:imsize[0]:8]:  
  for j in r_[:imsize[1]:8]:
    dct[i:(i+8),j:(j+8)] = scipy.fftpack.dct( scipy.fftpack.dct( im[i:(i+8),j:(j+8)], axis=0, norm='ortho' ), axis=1, norm='ortho' )
```

```python
#show DCT 
plt.figure()
plt.imshow(dct,vmax = np.max(dct)*0.01,vmin = 0)
print(dct)
```
